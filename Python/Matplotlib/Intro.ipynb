```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn  as sns

df = pd.read_csv('diamonds.csv')

df.head(10)
```

                                Explain each part of diamonds dataset:

                                A data frame with 53,940 rows and 10 variables:

                                • carat: Weight of the diamond (0.2--5.01)

                                • cut: Quality of the cut (Fair, Good, Very Good, Premium, Ideal)

                                • color: Diamond color, from D (best) to J (wpqt)

                                • clarity: A measurement of how clear the diHond is (II (worst), S12, SII, VS2, VSI, WS2, WSI, IF (best))

                                • depth: Total depth percentage = z/ mean(x, y) = 2 * z/ (x + y) (43--79)

                                • table: Width of top of diamond relative to widest point (43--95)

                                • price: Price in US dollars ($326--$18,823)

                                • x: Length in mm (0--10.74)

                                • y: Width in mm (0--58.9)

                                • z: Depth in mm (0--31.8)

```python
# Line plot
plt.plot(df['carat'],df['price'])
plt.title('Carat vs Price Line Plot')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
# Scatter plot
plt.scatter(df['carat'],df['price'])
plt.title('Carat vs Price Line Plot')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
# bar plot

plt.bar(df['cut'].unique(),df.groupby('cut')['price'].mean())
plt.title('Cut VS Average Price')
plt.xlabel('Cut')
plt.ylabel('Average Price')
plt.show()
```

```python
# intermediate ploting
# Histogram : it identifies pattern in data 

plt.hist(df['price'], bins=10, edgecolor = 'black')
plt.title('Distribution of price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# box plot : is a visual way to represent the distribution of a dataset

plt.boxplot(df['carat']) # to make it horizontal : vert = false
plt.title('Box Plot of Carat')
plt.xlabel('Carat')
plt.show()
```

```python
# pie chart : first convert the data into count 
cut_count = df['cut'].value_counts()
colors = ['skyblue', 'lightcoral', 'lightgreen', 'orange', 'lightpink'] 
explode = [0] * len(cut_count)  # Create a list of zeros
explode[0] = 0.1  # Explode the first slice (e.g., the largest)
plt.pie(cut_count, labels= cut_count.index, autopct='%1.1f%%', startangle=90, colors=colors, explode=explode)
plt.title('Distribution of cut quality')
plt.show()
```
